```python
quartieri_milano3857['Shape_Area'].mean()
```

```python
quartieri_maggiori_media = quartieri_milano3857[quartieri_milano3857['Shape_Area'] > quartieri_milano3857['Shape_Area'].mean()]
quartieri_maggiori_media
```

```python
ax = quartieri_maggiori_media.plot(figsize = (12,6), facecolor = 'None', edgecolor = 'k', linewidth = 2)
contextily.add_basemap(ax)
```

```python
ax = quartieri_maggiori_media.plot(figsize = (12,6), column = 'Shape_Area', legend = True, cmap = 'Reds', edgecolor = 'k')
contextily.add_basemap(ax)
```
