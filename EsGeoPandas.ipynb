## Troviamo il Galvo

```python
from shapely.geometry import Point 
galvani= Point(9.1885776, 45.5065988) # le coordinate sono da invertire, la seconda con la prima 
Quartieri[Quartieri.contains(galvani)]
```

```python
#stampare il quartieri del galvani 
Quartieri[Quartieri.contains(galvani)].plot()
```

```python
quartieregalvani = Quartieri[Quartieri.contains(galvani)]
quartieregalvani3857 = quartieregalvani.to_crs(epsg = 3857)
```

```python
#stampare il quartieri del galvani sulla mappa
ax = quartieregalvani3857.plot(figsize = (12,6), facecolor = 'none', edgecolor = 'k')
contextily.add_basemap(ax)
```

```python
from shapely.geometry import Point 
galvani = geopandas.GeoSeries([Point(9.1885776, 45.5065988)], crs = 4326)
galvani3857 = galvani.to_crs(epsg = 3857)
ax = quartieregalvani3857.plot(figsize = (12,6), facecolor = 'none', edgecolor = 'k')
galvani3857.plot(ax = ax)
contextily.add_basemap(ax)
```
