```python
quartieri = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/GeoPandas/ds964_nil_wm.zip')
quartieri
```

```python
quartieri.plot(figsize = (12,6),edgecolor = 'Blue', facecolor = 'None')
```

cercare il quartiere che contiene il Duomo di Milano

```python
from shapely.geometry import Point
duomo = Point(9.191926, 45.464098)
quartiere_duomo = quartieri[quartieri.contains(duomo)] #trovare il quartiere che contiene il duomo, quando usiamo le proprietà es. contains usaimo .item
quartiere_duomo.plot()
duomo_serie = geopandas.GeoSeries([duomo], crs = 4326) # 4326 è il crs dei quartieri
duomo_serie3857 = duomo_serie.to_crs(epsg = 3857)
```

```python
quartiere_duomo.crs
```

```python
quartiere_duomo3857 = quartiere_duomo.to_crs(epsg = 3857)
```

```python
ax = quartiere_duomo3857.plot(facecolor = 'None', edgecolor = 'k')
duomo_serie3857.plot(ax = ax, markersize = 20)
contextily.add_basemap(ax)
```

```python
quartieri_confinanti_duomo = quartieri[quartieri.touches(quartiere_duomo)] #abbiamo trovato i quartieri confinanti al duomo 
quartieri_confinanti_duomo.plot()
```
