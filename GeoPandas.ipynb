```python
%matplotlib inline

import pandas as pd
import geopandas
```

```python
countries = geopandas.read_file("https://github.com/jorisvandenbossche/geopandas-tutorial/blob/main/data/ne_110m_admin_0_countries.zip?raw=true")
countries
```

```python
countries.plot()
# asse x longitudine da -180 a +180 (asse orizzontale)
# asse y altitudine da -90 a +90 (asse verticale)
```

```python
#prende tutte le nazioni africane
africa = countries[countries['continent'] == 'Africa']
africa
```

```python
africa.plot()
```

```python
countries.geometry #prende solo la colonna geometria 
```

```python
countries[countries['name'] == 'Italy'].area #calcolo l'area
```

```python
cities = geopandas.read_file("https://github.com/jorisvandenbossche/geopandas-tutorial/blob/main/data/ne_110m_populated_places.zip?raw=true")
cities
```

```python
ax = countries.plot(edgecolor='k', facecolor='none', figsize=(15, 10))
cities.plot(ax=ax, color='red')
ax.set(xlim=(-20, 60), ylim=(-40, 40)) #coordinate che ti permettono di vedere l'Africa
```

```python
ax = countries.plot(edgecolor='k', facecolor='none', figsize=(15, 10))
cities.plot(ax=ax, color='red')
ax.set(xlim=(5, 20), ylim=(35, 48)) 
```

```python
stations = geopandas.read_file('https://github.com/jorisvandenbossche/geopandas-tutorial/blob/main/data/paris_bike_stations_mercator.gpkg?raw=true')
stations
```

```python
import contextily
```

```python
ax = stations.plot(figsize=(12,6), markersize=5)
contextily.add_basemap(ax)
```

```python
#linea orizzontale = posti, linea verticale numero di stazioni
stations['bike_stands'].hist() #hist = istogramma
```

```python
ax = stations.plot(figsize=(15, 10), column='available_bikes', legend=True,  markersize=5) 
contextily.add_basemap(ax)
```

```python
districts = geopandas.read_file("https://raw.githubusercontent.com/jorisvandenbossche/geopandas-tutorial/main/data/paris_districts_utm.geojson")
districts
```

```python
districts.plot()
```

```python
ax = districts.plot()
contextily.add_basemap(ax)
```

```python
districts.geometry.area
```

```python
districts['area'] = districts.geometry.area / 1e6
districts
```

```python
districts['population_density'] = districts['population'] / districts.geometry.area * 10**6
districts
```

```python
districts.plot(column='population_density', figsize=(12, 6), legend=True)
```
